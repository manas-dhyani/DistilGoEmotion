```python
# Cell: Build a CSV listing waveform paths and labels (modify label extraction logic per your dataset layout)
import os
import pandas as pd
root = os.path.join(DRIVE_ROOT, 'archive') # adjust if needed
wav_paths = []
for dp, dn, filenames in os.walk(root):
  for f in filenames:
    if f.lower().endswith('.wav'):
      wav_paths.append(os.path.join(dp, f))
len(wav_paths)


# Example: CREMA-D filenames often contain emotion labels, e.g. '101_RAB_NEU.wav' or similar.
# Update the parsing logic below to match your filenames.


rows = []
for p in wav_paths:
  fname = os.path.basename(p)
# Example parsing: tokens separated by '_' and emotion code at index 2 or 3
# Adapt this to your dataset's naming convention
  parts = fname.split('_')
# naive fallback
  emotion = 'unknown'
  for tok in parts:
    tok_low = tok.lower()
    if tok_low.startswith('ang') or tok_low.startswith('angry') or tok_low.startswith('ang'):
      emotion = 'anger'
    if tok_low.startswith('hap') or 'happy' in tok_low:
      emotion = 'happy'
    if tok_low.startswith('neu'):
      emotion = 'neutral'
    if tok_low.startswith('sad'):
      emotion = 'sad'
    if tok_low.startswith('dis') or 'disgust' in tok_low:
      emotion = 'disgust'
    if tok_low.startswith('fear') or 'fear' in tok_low:
      emotion = 'fear'
  rows.append({'path': p, 'emotion': emotion})


meta = pd.DataFrame(rows)
meta['emotion'].value_counts()
meta.to_csv(os.path.join(DRIVE_ROOT, 'cremad_metadata.csv'), index=False)
print('metadata saved')
```

```python
# 🧱 Cell: Build Hugging Face Dataset
from datasets import Dataset, Audio
import pandas as pd
import os
import librosa
from datasets import ClassLabel, Features, Value
DRIVE_ROOT = '/content/drive/MyDrive/wav2vec_cremad'

# Load metadata CSV
meta = pd.read_csv(os.path.join(DRIVE_ROOT, 'cremad_metadata.csv'))

# Keep only known labels
meta = meta[meta.emotion != 'unknown'].reset_index(drop=True)
label_list = sorted(meta.emotion.unique())
label2id = {l: i for i, l in enumerate(label_list)}

# Create HF dataset
hf_ds = Dataset.from_pandas(meta)

# Cast the 'path' column to Audio type (for automatic loading)
hf_ds = hf_ds.cast_column('path', Audio(sampling_rate=16000))

# Map emotion labels to integer IDs
def map_labels(batch):
    batch['label'] = label2id[batch['emotion']]
    return batch

hf_ds = hf_ds.map(map_labels)

# Quick train/test split (15% test set)

# Define label feature
label_feature = ClassLabel(names=label_list)

# Recast dataset features
hf_ds = hf_ds.cast(
    Features({
        'path': Audio(sampling_rate=16000),
        'emotion': Value('string'),
        'label': label_feature,
    })
)

# Now you can stratify safely
hf_ds = hf_ds.train_test_split(test_size=0.15, stratify_by_column='label', seed=42)

# Display a small sample
hf_ds
```

```python
# 🎧 Cell: Feature Extractor + Data Collator
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2ForSequenceClassification, DataCollatorWithPadding
from dataclasses import dataclass
import numpy as np
import torch

# Model & feature extractor
model_name = 'facebook/wav2vec2-base'
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained(model_name, sampling_rate=16000)

# 🧩 Preprocessing function
def preprocess(batch):
    # Extract the audio array
    speech_array = batch["path"]["array"]

    # Ensure float32 numpy array
    if speech_array.dtype != "float32":
        speech_array = speech_array.astype("float32")

    # Always pass a *list* to get consistent dict output
    features = feature_extractor(
        [speech_array],
        sampling_rate=16000,
        padding=True,
        return_tensors="np"
    )

    batch["input_values"] = features["input_values"][0]
    # Some feature extractors may not include attention_mask
    batch["attention_mask"] = (
        features["attention_mask"][0]
        if "attention_mask" in features
        else np.ones_like(features["input_values"][0], dtype=np.int64)
    )

    return batch


# Apply preprocessing
hf_ds = hf_ds.map(preprocess, remove_columns=['path', 'emotion'])

# 🧱 Data Collator
@dataclass
class DataCollatorCTCWithPadding:
    feature_extractor: Wav2Vec2FeatureExtractor
    padding: bool = True

    def __call__(self, features):
        input_values = [f['input_values'] for f in features]
        labels = [f['label'] for f in features]

        batch = self.feature_extractor.pad(
            {'input_values': input_values},
            return_tensors='pt'
        )
        batch['labels'] = torch.tensor(labels)
        return batch

# Instantiate collator
collator = DataCollatorCTCWithPadding(feature_extractor=feature_extractor)
```

```python
# Cell: Model, training args, trainer

import os
import torch
from transformers import TrainingArguments, Trainer, Wav2Vec2ForSequenceClassification
from sklearn.metrics import accuracy_score, f1_score

# Initialize model
num_labels = len(label_list)
model = Wav2Vec2ForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_labels
)

# Define training arguments
training_args = TrainingArguments(
    output_dir=os.path.join(DRIVE_ROOT, "wav2vec2-finetuned"),
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    eval_strategy="epoch",
    save_strategy="epoch",
    logging_steps=50,
    fp16=torch.cuda.is_available(),
    num_train_epochs=6,
    learning_rate=1e-4,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
)

# Compute metrics function
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average="weighted")
    return {"accuracy": acc, "f1": f1}

# Initialize trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=hf_ds["train"],
    eval_dataset=hf_ds["test"],
    data_collator=collator,
    compute_metrics=compute_metrics,
)

# Train model
trainer.train()

# Save final model + feature extractor
final_path = os.path.join(DRIVE_ROOT, "wav2vec2-finetuned-final")
trainer.save_model(final_path)
feature_extractor.save_pretrained(final_path)

print(f"✅ Model and feature extractor saved to: {final_path}")
```
